## Draw multiple graphs in one single figure( Subplots)

```python
rcParams['figure.figsize']=8,8
```

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].pie(data)
ax[1,0].scatter(x,y)
ax[1,1].bar(x,y)
plt.show()
```
