## Data Augmentation

```python
#import necessary libraries

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# data augmentation for training variable

train_datagen = ImageDataGenerator(rescale = 1./255,zoom_range = 0.2,horizontal_flip=True)
```

```python
# data augmentation for testing variable

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# data augmentation on the training data

xtrain = train_datagen.flow_from_directory('/content/dataset/Training',
                                           target_size =(64,64),
                                           class_mode = 'categorical',
                                           batch_size = 100)
```

```python
xtest = test_datagen.flow_from_directory('/content/dataset/Testing',
                                           target_size =(64,64),
                                           class_mode = 'categorical',
                                           batch_size = 100)
```

## CNN Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D,MaxPooling2D,Flatten,Dense
```

```python
## adding layers

model = Sequential() # Initializing sequential model
model.add(Convolution2D(32,(3,3),activation='relu',input_shape=(64,64,3))) # convolution layer
model.add(MaxPooling2D(pool_size=(2, 2))) # Max pooling layer
model.add(Flatten()) # Flatten layer
model.add(Dense(300,activation='relu')) # Hidden layer 1
model.add(Dense(150,activation='relu')) # Hidden layer 2
model.add(Dense(4,activation='softmax')) # Output layer
```

```python
# Compile the model

model.compile(optimizer = 'adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
# Train the model

model.fit_generator(xtrain,steps_per_epoch=len(xtrain),
                    epochs=10,validation_data=xtest,validation_steps=len(xtest))
```

```python
# Save model

model.save('ibm_animal3.h5')


!tar -zcvf ibm3_animal-classification.tgz ibm_animal3.h5
```
