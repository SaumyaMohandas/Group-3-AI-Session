```python
## Import the required libraries

import numpy as np
import pandas as pd
```

```python
# reading the dataset

df=pd.read_csv('/content/50_Startups.csv')
df.head()
```

```python
df.shape
```

```python
# checking null values

df.isnull().sum()
```

```python
# checking the data type

df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# initializing
le=LabelEncoder()
```

```python
df['State'] = le.fit_transform(df['State'])
```

```python
df.head()
```

```python
# split the data(independent and dependent)

x= df.iloc[:,:-1].values
y = df.iloc[:,4:5].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting the data(Training & testing)
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size= 0.3,random_state = 0)
```

```python
# finding the shape of train and test data

xtrain.shape, xtest.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# creating ANN skeleton 

regressor = Sequential()
regressor.add(Dense(4,activation='relu'))
regressor.add(Dense(12,activation='relu'))
regressor.add(Dense(8,activation='relu'))
regressor.add(Dense(9,activation='relu'))
regressor.add(Dense(1,activation='linear'))
```

```python
# compiling the ANN model

regressor.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
# trainig the model
regressor.fit(xtrain,ytrain,batch_size = 10,epochs=300)
```

```python
# Testing the model
ypred = regressor.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,ypred)*100
```

```python
ypred.flatten()
```

```python
pd.DataFrame({'Actual_value':ytest.flatten(),'predicted_value':ypred.flatten()}).head(10)
```
