## Merge

```python
dd= pd.DataFrame({
    'name':['Appu','Ammu','Anil','Aswin'],
    'Maths':[78,73,77,75]})
```

```python
dd
```

```python
de= pd.DataFrame({
    'name':['Appu','Ammu','Anil','Aswin'],
    'English':[88,84,87,89]})
```

```python
de
```

```python
Merge= pd.merge(dd,de, on='name')
```

```python
Merge
```

```python
ds= pd.DataFrame({
    'name':['Appu','Ammu','Anil','Aswin'],
    'science':[80,84,81,85]})
```

```python
M1= pd.merge(Merge,ds, on='name')
M1
```
