# Face Detection

```python
img = cv2.imread('human.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Initializing the path to variable
haar_cas = 'haarcascade_frontalface_alt.xml'
haar_cas
```

```python
# Reading XML file
face_cas = cv2.CascadeClassifier(haar_cas)
face_cas
```

```python
# Detect face

faces = face_cas.detectMultiScale(img,scaleFactor=1.1,minNeighbors=5)
faces
```

```python
print('{} faces detected!!!'.format(len(faces)))
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,200,180),5)

plt.imshow(img)
```

# Number plate Detection

```python
img = cv2.imread('number_plate.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
haar_cas = 'haarcascade_russian_plate_number.xml'
num_cas = cv2.CascadeClassifier(haar_cas)
num_plate = num_cas.detectMultiScale(img,scaleFactor=1.1,minNeighbors=5)
print('{} number plate detected!!!'.format(len(num_plate)))
for (x,y,w,h) in num_plate:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)

plt.imshow(img)
```
